```python
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_community.tools import tool
import requests
```

```python
# duckduckgo search tool
search_tool = DuckDuckGoSearchRun()

response = search_tool.invoke("Top news in India today")
print(response)
```

```python
# create llm
from langchain_groq import ChatGroq

llm = ChatGroq(model="llama-3.3-70b-versatile")
llm.invoke("Hey")
```

Now that we have a tool, and a llm, we can create an agent.

```python
from langchain_classic.agents import AgentExecutor,create_react_agent
from langchain_classic import hub

# pull ReAct prompt from langchain hub
prompt = hub.pull("hwchase17/react") 
# pull the standard ReAct agent prompt from langchain hub
# we are building a react agent --> reasoning + action
```

```python
# now create the ReAct agent using the prompt, the tools and the llm

agent = create_react_agent(
    llm=llm,
    tools=[search_tool],
    prompt=prompt
)

# AgentExecutor is the guy who performs the tasks, (executes) the tasks
agent_executor = AgentExecutor(
    agent=agent,
    tools=[search_tool],
    verbose=True
)
```

```python
# so now invoke the agent_executor
response = agent_executor.invoke({"input":"Tell me top 5 places to visit in Kolkata"})
print(response)
```
