```python
from langchain_community.retrievers import WikipediaRetriever
```

```python
# initialize the retriever
retriever = WikipediaRetriever(
    top_k_results=3, # how many top results you want
    lang='en' 
)

# define the query
query = "What is Pakistan occupied Kashmir?"

# send the query to the retriever and fetch the top k similar docs
docs = retriever.invoke(query)
docs
```

```python
# print the documents one by one
for i,doc in enumerate(docs):
    print(f"Result {i+1} :")
    print(doc.page_content)
```
