```python
from langchain_chroma import Chroma
from langchain_core.documents import Document
from langchain_ollama import OllamaEmbeddings
```

```python
# create documents

doc1 = Document(
    page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history. Known for his aggressive batting style and strong leadership skills.",
    metadata = {"team":"Royal Challengers Bangalore"}
)

doc2 = Document(
    page_content="Rohit Sharma is the most successful captain in IPL history, leading Mumbai Indians to five titles. He's known for his calm leadership and explosive batting.",
    metadata={"team": "Mumbai Indians"}
)

doc3 = Document(
    page_content="MS Dhoni, famously known as Captain Cool, has led Chennai Super Kings to multiple IPL titles. His finishing skills, wicketkeeping, and calm decision-making are legendary.",
    metadata={"team": "Chennai Super Kings"}
)

doc4 = Document(
    page_content="Jasprit Bumrah is considered one of the best fast bowlers in T20 cricket. Playing for Mumbai Indians, he is known for his deadly yorkers and consistency.",
    metadata={"team": "Mumbai Indians"}
)

doc5 = Document(
    page_content="Ravindra Jadeja is a dynamic all-rounder who contributes with both bat and ball. Representing Chennai Super Kings, his quick fielding and match-winning performances stand out.",
    metadata={"team": "Chennai Super Kings"}
)
```

```python
# create an embedding model
embedding = OllamaEmbeddings(
    model="nomic-embed-text",
)

# and take all the docs in a list
docs = [doc1,doc2,doc3,doc4,doc5]
```

```python
embedding.embed_query("Aadrish")
```

```python
# define the vector store where you will store the embeddings
vector_store = Chroma(
    embedding_function=embedding,
    persist_directory='chroma_db', # store the vectors in a sqlite3 file named chroma_db
    collection_name="sample" # the vectors will be stored in a collection named "sample"
)
```

```python
# now add the documents to the vector store
vector_store.add_documents(documents=docs)

# and every document will be assigned a unique id
```

Every document has been assigned an unique ID.

```python
# view the documents in your vector DB
# you can choose what you want to view from each document in the 'include' attribute
vector_store.get(include=['embeddings','documents','metadatas']) # here we view the embeddings, the actual documents and the metadatas for each document
```

#### Similarity Search

```python
# search top k similar vectors from the vector_store , wrt a query
vector_store.similarity_search(
    query="Who is a good bowler?",
    k=2
)
```

```python
# search with similarity score, these scores represent a distance, so lesser = better
vector_store.similarity_search_with_score(
    query="Who among them is a good bowler?",
    k=2
)
```

```python
# meta-data filtering
# I want to get the documents, on Chennai Super Kings
vector_store.similarity_search(
    query="",
    filter={"team":"Chennai Super Kings"}
)
```

```python
# update document
updated_doc1 = Document(
    page_content="Virat Kohli, the former captain of Royal Challengers Bangalore, is known for his aggressive batting style and strong leadership skills.",
    metadata={"team":"Royal Challengers Bangalore"}
)
```

```python
# update the document in the vector store
vector_store.update_document(document_id="2453b20e-65b8-4410-9193-a4d1ee2d51dc",document=updated_doc1)
```

```python
# now view the updated document
vector_store.get(include=["embeddings","documents","metadatas"])
```
