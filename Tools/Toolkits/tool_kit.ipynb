```python
from langchain_community.tools import tool
```

```python
# create custom tools
@tool
def add(a:int,b:int)->int:
    """Adds two integers"""
    return a+b

@tool
def multiply(a:int,b:int):
    """Multiplies two integers"""
    return a*b
```

```python
# create a MathToolkit class
class MathToolkit:
    def get_tools(self):
        return [add,multiply]
```

```python
# create an instance of the the toolkit class
toolkit = MathToolkit()
tools = toolkit.get_tools()
```

```python
for tool in tools:
    print(tool.name,'=>',tool.description)
```
