#### DuckDuckGo search tool

```python
from langchain_community.tools import DuckDuckGoSearchRun

# search tool
search_tool = DuckDuckGoSearchRun()

# tools are also runnables
result = search_tool.invoke("NVIDIA")

print(result)
```

#### Shell tool

```python
from langchain_community.tools import ShellTool

# shell tool
shell_tool = ShellTool()

result = shell_tool.invoke("whoami")
print(result)
```

#### WikipediaQueryRun Tool

```python
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper

wiki_api = WikipediaAPIWrapper()

# wikipedia query run tool
wikipedia_tool = WikipediaQueryRun(api_wrapper=wiki_api)

result = wikipedia_tool.invoke("Shinchan")
print(result)
```
