```python
from langchain_community.tools import StructuredTool
from pydantic import BaseModel,Field
```

```python
# create a pydantic schema
class MultiplyNumbers(BaseModel):
    a : int = Field(required=True,description="The first number to multiply")
    b : int = Field(required=True,description="The second number to multiply")
```

```python
# now create the actual tool function
def multiply(a:int,b:int)->int:
    return a*b

# convert the function into a tool, using structured tool
multiply_tool = StructuredTool.from_function(
    func=multiply,
    name="multiply",
    description="A tool that multiplies two numbers",
    args_schema=MultiplyNumbers
)
```

```python
# now test the tool
multiply_tool.invoke({"a":5,"b":7})
```

```python
# view details about the tool
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```
