```python
from langchain_community.tools import tool
```

```python
# create a custom tool in 3 steps 

# Step 1 --> create the function with docstring
def multiply(a,b):
    """Multiply two numbers"""
    return a*b

# Step 2 --> create the function, with type hinting, this will enable the llm to understand the i/p dtype and o/p dtype
def multiply(a:int,b:int) -> int:
    """Multiply two numbers"""
    return a*b

# Step 3 --> add tool decorator to the function, which enables and llm to interact with this function
@tool
def multiply(a:int,b:int)->int:
    """Multiply two numbers"""
    return a*b
```

Now the function multiply is a tool, and an llm can interact with it.

```python
# since a tool is a runnable, we can invoke it
multiply.invoke({"a":2,"b":4})
```

```python
# see the name of the tool
print(multiply.name)
```

```python
# description of the tool
print(multiply.description)
```

```python
# and the args the tool takes input
print(multiply.args)
```

```python
# and when llm calls this tool, the llm sees this schema for the tool
print(multiply.args_schema.model_json_schema())
```
